# Intervalos multi-dimensionales 

Para resolver problemas en más que una dimensión, será necesario poder calcular un *enclosure* ("encierro") de la imagen de una función sobre un conjunto en varias dimensiones.
Para ello, podemos definir **intervalos multi-dimensionales**, de manera análoga a los intervalos uni-dimensionales que ya conocemos.

[1] ¿Cuál sería una definición matemática razonable de un intervalo multi-dimensional?

```julia
x=Intervalo[]
divisiones=20
dx=1/divisiones
for i in 1:divisiones
    I=Intervalo((i-1)*dx, dx*i) 
    push!(x,I)
end

a= Enclosure[]
for j in 1:divisiones
    Ij=x[j]
    for i in 1:divisiones
        Ii=x[i]
        e={Ij, Ii}
        push!(a, e)
    end
end
```

[5] Haz lo mismo para la imagen del cuadrado unitario bajo la función 

$$ \begin{pmatrix} x \\ y \end{pmatrix} \mapsto 
\begin{pmatrix} 1 - ax^2 + y \\ b x \end{pmatrix},$$

con $a = 1.4$ y $b=0.3$ (el famoso "mapeo de Hénon").

```julia
henon(x)={1-1.4*x[1]^2+x[2], 0.3*x[1]}
```

```julia
using PyPlot

PyPlot.fill_between([0,2],0,2,color="blue", alpha=0.00001)

for i in 1:length(a)
    m=henon(a[i])
    PyPlot.fill_between([float64(m[1].a),float64(m[1].b)],float64(m[2].a),float64(m[2].b),color="blue", alpha=0.2)
end
```

que es una parte del atractor de henon. Ahora voy a graficar una sección mas grande y tomando una división mas detallada.

```julia
x=Intervalo[]
divisiones=200
dx=10/divisiones
for i in 1:divisiones
    I=Intervalo((i-1)*dx-5, dx*i-5) 
    push!(x,I)
end

a= Enclosure[]
for j in 1:divisiones
    Ij=x[j]
    for i in 1:divisiones
        Ii=x[i]
        e={Ij, Ii}
        push!(a, e)
    end
end
```

```julia
using PyPlot


PyPlot.fill_between([-1,1],-1,1,color="blue", alpha=0.00001)

for i in 1:length(a)
    m=henon(a[i])
    PyPlot.fill_between([float64(m[1].a),float64(m[1].b)],float64(m[2].a),float64(m[2].b),color="blue", alpha=0.2)
end
```

Es interesante el hecho de  que  todos los puntos contenidos en la zona [-5, 5]x[-5,5] son enviados a la zona marcada  por el atractor de Henon.
