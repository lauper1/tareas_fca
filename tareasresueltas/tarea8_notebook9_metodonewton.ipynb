# El método de Newton

Recordemos que el *método de Newton* es un método numérico *iterativo* para encontrar raíces de funciones (continuamente) diferenciables.

Sea $f$ la función cuyas raíces queremos encontrar.
La idea es que empecemos desde una adivinanza inicial $x_0$, y que la siguiente aproximación $x_1$ esté donde la recta tangente a la curva $f(x)$ en $x_0$ corte el eje $x$. 

[1] Esboza la geometría, tanto a mano como en la computadora, y así encuentra la expresión de $x_{n+1}$ en términos de $x_n$.

La tangente a la gráfica de la función en el punto $x_0$ tiene pendiente  $f'(x_0)$. Para encontrar la ordenada al origen observamos que debe pasar por el punto $(x_0,f(x_0))$. De modo que la recta tangente esta dada por:
$$tangente_{x_0}(x)=f'(x_0)x+ f(x_0)-f'(x_0)x_0$$

Resolviendo para el punto en que corta al eje x. Da

$$x=x_0-f(x_0)/f'(x_0)$$



```julia
using PyPlot
f(x)=x^2-1#grafica(a, b, f::Function)
df(x)=2x
x = [-5:0.1:5];
y=[f(x[i]) for i = 1:length(x)];
plot(x,y, color="purple")
x0=-3
tan(x)= df(x0)*(x-x0)+f(x0)
u=[tan(x[i]) for i = 1:length(x)];
cero=[0 for i=1:length(x)]
plot(x,u)# "-", color="purple")
plot(x,cero)
x0=x0-f(x0)/df(x0)
tan1(x)= df(x0)*(x-x0)+f(x0)
u1=[tan1(x[i]) for i = 1:length(x)];
plot(x,u1)
```

 de modo que podemos realizar el proceso iterativo
 $$x_{n+1}=x_n-f(x_n)/f'(x_n)$$

[6] Utiliza tu algoritmo para calcular raíces de funciones conocidas multidimensionales.

Para funciones que van de Rn→Rn,  el Jacobiano está dado por:
$$J_{ij}=\frac{∂F_i}{∂x_j}$$

y los puntos sucesivos para hallar las raíces por medio del Jacobiano cumoplen  la ecuación:
$$J(x_n)(x_{n+1}−x_n)=F(x_n),$$
de donde:
$$x_{n+1}=x_n+J^{-1}(x_n)⋅F(x_n).$$

Primero defino el Jacobiano.

```julia
using Derivadasa

function jacobiano(x::Array, f::Function)  
    m=length(x)
    n=length(f(x))
    J=zeros(n,m)
    for i=1:n
        for j=1:m
            xvar=fdev[]
            for k=1:m
                if k==j
                    comp=fdev(x[k],1.0)
                else
                    comp=fdev(x[k])
                end
                xvar=push!(xvar,comp)
            end   
            J[i,j]=f(xvar)[i].d
        end
    end
    J
end
```

vamos a probarlo con algunas funciones

$$f(x,y)= [x^2-7yx-9, x^5-y^3]$$

```julia
function F(x)
    [x[1]^2-7x[2]-9, x[1]-x[2]]
end
y=[1,2.9]
```

```julia
function newton(x::Array,F::Function, lim)
    dif=lim+1
    n=length(x)
    k=0 #contador por si necesito saber cuantas iteraciones hizo
    while dif>lim
            J=jacobiano(x,F)
            x0=x
            x=x0-(J^-1)*F(x0)
            dif=norm(x0-x)/norm(x0)
        k=k+1
        end
        x
end
```
